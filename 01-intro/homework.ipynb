```python
df_jan = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet')

print(f'The January dataset has {df_jan.shape[1]} columns.')
```

## Q2. Computing duration

Now let's compute the `duration` variable. It should contain the duration of a ride in minutes. 

What's the standard deviation of the trips duration in January?

* 32.59
* 42.59
* 52.59
* 62.59

```python
df_jan_processed = df_jan.copy()
df_jan_processed['duration'] = df_jan_processed['tpep_dropoff_datetime'] - df_jan_processed['tpep_pickup_datetime'] 
df_jan_processed['duration'] = df_jan_processed['duration'].apply(lambda td: td.total_seconds() / 60)
# df_jan_processed.describe().T
std_duration = df_jan_processed['duration'].std()
print(f'The standard deviation of trips duration in January was {std_duration:.2f} seconds.')
```
